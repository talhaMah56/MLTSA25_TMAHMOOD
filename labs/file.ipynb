```python
import numpy as np
np.random.seed(42)
import matplotlib.pyplot as plt
```

```python
N = 100
y = np.random.randn(N)
plt.hist(y)
```

```python
t = np.linspace(0, N, N)
t.shape, y.shape
```

```python
plt.plot(t, y)
plt.xlabel("t")
plt.ylabel("y")
```
